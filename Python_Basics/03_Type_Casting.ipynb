### Type Casting: ###

```python
- Change one Data type to another data type 
- int ---> float
- int ---> string
- int ---> boolean
- int ---> complex

- similarly float to others 
- string to others
```

**integer to other data types**

```python
num1=100
print(num1)
type(num1)
```

```python
# case-1: int --> float
float_num=float(num1)  # num1 converted into float
float_num
```

```python
type(float_num)
```

```python
# case-2: int to string
string_num=str(num1)
string_num
```

```python
type(string_num)
```

```python
# Case-3: integer to boolean
boolean_num=bool(num1)
boolean_num
```

```python
type(boolean_num)
```

```python
bool(100)    # when it will give false 
# only give "false"=bool(0) for 0, for other numbers "true"
```

```python
# case-4: integer to complex 
complex_num=complex(num1)
complex_num
```

```python
complex_num=complex(num1,num1)
complex_num
```

-  num1=100
-  float_num=float(num1)
-  string_num=str(num1)
-  boolean_num=bool(num1)
-  complex_num=complex(num1)
