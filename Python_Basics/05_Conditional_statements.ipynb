```python
# using try & exception block
try:
    num=eval(input("enter a number:"))
    if num%2==0:
        print("{} is a even number".format(num))
    else:
        print("{} is a odd number".format(num))

except Exception as error:
    print(error)
```

```python
# WAP ask the user to enter percentage of marks
# if per >90 pritn A
# if per between 70 ot 90 print B
# if per between 50 to70 print C
# otherwise print D
# if - elif - elif- else


per=eval(input("enter a percentage:"))
if per>90:
    print("{} percentage belongs to 'A' grade".format(per))

elif 70<per<=90:
    print("{} percentage belongs to 'B' grade".format(per))

elif 50<per<=70:
    print("{} percentage belongs to 'C' grade".format(per))

else:
    print("{} percentage belongs to 'D' grade".format(per))


```

```python
# WAP ask the user enter age                                            # 14 th class
# if age > 60 print (SC)
# if age is between 45 to 60 (aged)
# if age is between 30 to 45 (MA)
# if age is between 20 to 30 (young)
# if age is between 13 to 20 (teenage)
# otherwise Child


age=eval(input("enter a age of the person:"))

if age>=60:
    print("Senior citizen")

elif age>=45:
    print("aged person")

elif age>=30:
    print("medium age ")

elif age>=20:
    print("young age")

elif age>=13:
    print("teen age")
    
else:
    print("child")
```

```python
# WAP ask the user enter distance in KM                                                               "15 th class" 
# if dis > 10 km :print charge is 100 rs                                          
# if distance is betwn 7 to 10 km : print charge is 70 rs
# if dis is btwn 4 to 7 km : charge is 40 rs
# otherwise print charge is 30 rs
# total 4: print charge is 30 rs



dis=eval(input("enter the distance in Km:"))

if dis>=10:
    print(f"for {dis} Km charge is 100 rs")

elif dis>=7:
    print(f"for {dis} Km charge is 70 rs")

elif dis>=4:
    print(f"for {dis} Km charge is 40 rs")

else:
    print(f"for {dis} Km charge is 30 rs")
```

```python
# WAP ask user to enter salary                                                                       "15 th class" 
# if salary > 10>lakhs
#         enter the user  tax percentage : 10%
#         print: the amount of tax paied : (salary*tax_per/100)
# if salary bet 7 lakhs to 10 lakhs
#         ask the user enter tax percentage: 5%
#         print the amount of tax paid
# if salary btween 4 to 7 lakhs
#         ask the user enter tax percentage: 2.5%
#         print the amount of tax paid
# otherwise: print("no tax required")


sal=eval(input("enter salary:"))

if sal>=1000000:
    tax=eval(input(f"enter the tax for {sal} lakhs in percentage"))
    tax_paid=(sal*tax)/100
    print(f"total tax paid for {sal} is {tax_paid} rs")
    

elif sal>=700000:
    tax=eval(input(f"enter the tax for {sal} lakhs in percentage"))
    tax_paid=(sal*tax)/100
    print(f"total tax paid for {sal} is {tax_paid} ")

elif sal>=400000:
    tax=eval(input(f"enter the tax for {sal} lakhs in percentage"))
    tax_paid=(sal*tax)/100
    print(f"total tax paid for {sal} is {tax_paid} rs")

else:
    print(f"tax is not required for {sal}")
```
